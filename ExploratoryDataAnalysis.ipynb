# Using Axes:

```python
#Multiple plot in same axis
z=np.cos(x)
fig, (ax1, ax2) = plt.subplots( 2,1,sharex=True, figsize=(12, 6))

# Plot sine wave on the first subplot
ax1.plot(x, y, label='Sine Wave', color='b', linestyle='-', linewidth=2)
ax1.set_title('Sine Wave')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.grid(True)
ax1.legend()

# Plot cosine wave on the second subplot
ax2.plot(x, z, label='Cosine Wave', color='r', linestyle='--', linewidth=2)
ax2.set_title('Cosine Wave')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.grid(True)
ax2.legend()
plt.show()
```

```python
#Multiple plot in different axis
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(12, 6))

# Plot sine wave on the first subplot
ax1.plot(x, y, label='Sine Wave', color='b', linestyle='-', linewidth=2)
ax1.set_title('Sine Wave')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.grid(True)
ax1.legend()

# Plot cosine wave on the second subplot
ax2.plot(x, z, label='Cosine Wave', color='r', linestyle='--', linewidth=2)
ax2.set_title('Cosine Wave')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.grid(True)
ax2.legend()
plt.show()
```

```python
y = np.sin(x)
z = np.cos(x)
a = np.tan(x)
b = 1 / np.tan(x)  # cotangent is 1/tangent

# Create subplots with a 2x2 layout
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 12))

# Plot sine wave on the first subplot
ax1.plot(x, y, label='Sine Wave', color='b', linestyle='-', linewidth=2)
ax1.set_title('Sine Wave')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.grid(True)
ax1.legend()

# Plot cosine wave on the second subplot
ax2.plot(x, z, label='Cosine Wave', color='r', linestyle='--', linewidth=2)
ax2.set_title('Cosine Wave')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.grid(True)
ax2.legend()

# Plot tangent wave on the third subplot
ax3.plot(x, a, label='Tangent Wave', color='g', linestyle='-.', linewidth=2)
ax3.set_title('Tangent Wave')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.grid(True)
ax3.legend()

# Plot cotangent wave on the fourth subplot
ax4.plot(x, b, label='Cotangent Wave', color='m', linestyle=':', linewidth=2)
ax4.set_title('Cotangent Wave')
ax4.set_xlabel('X')
ax4.set_ylabel('Y')
ax4.grid(True)
ax4.legend()

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

# Feature selection and engineering

Using p-score Use SelectKBest, f_regression from sklearn

```python
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.impute import SimpleImputer  # Import SimpleImputer
```

```python
from sklearn.impute import SimpleImputer  # Import SimpleImputer

# Load the mpg dataset
mpg = sns.load_dataset('mpg')

# Separate features (X) and target (y)
X = mpg.drop(columns=['mpg', 'origin', 'name'])  # Drop non-numeric columns
y = mpg['mpg']

# Impute missing values (replace NaNs with mean of the column)
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Convert back to DataFrame (optional)
X_imputed = pd.DataFrame(X_imputed, columns=X.columns)

# Perform train-test split on imputed data
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Initialize SelectKBest with f_regression
selector = SelectKBest(score_func=f_regression, k=3)  # Select top 3 features

# Fit selector to training data
X_train_selected = selector.fit_transform(X_train, y_train)

# Get columns selected by SelectKBest
selected_features = X.columns[selector.get_support()]

# Print selected features
print("Selected features based on f_regression scores:")
print(selected_features)
```
